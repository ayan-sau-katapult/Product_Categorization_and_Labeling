```python
import glob
import pandas as pd

# get data file names
path =r'C:\Users\AyanSau\Desktop\ABESofMAINE_ALLparsedFILES\ABESofMAINE_All_Files'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_df = pd.concat(dfs, ignore_index=True)
big_df
```

```python
big_df.isnull().sum()
```

```python
big_df["name"]=big_df["name"].fillna(big_df["Name"])
big_df["price"]=big_df["price"].fillna(big_df["Price"])
```

```python
big_df.isnull().sum()
```

```python
big_df=big_df[["name","price","Category","Leasable"]]
big_df
```

```python
big_df.to_excel("big_df.xlsx")
```

```python
def categorize(string):
    
    string= "".join(str(string).split()).lower()
    if "mouse" in string or "keyboard" in string:
        return "Desktop Accessories"
    elif "cooktop" in string or "glass" in string or "steel" in string:
        return "CookWare"
    elif "oven" in string or "grill" in string or "washer" in string or "microwave" in string:
        return "Home Appliances"
    elif "port" in string or "router" in string or "cable" in string or "ethernet" in string or "switch" in string or "transceiver" in string or "swtch" in string:
        return "Routers and Cables"
    elif "video" in string or "cam" in string and "len" not in string and "case" not in string and "bag" not in string and "pouch" not in string:
        return "Cameras"
    elif "batter" in string:
        return "Batteries"
    elif "filter" in string or "lens" in string or "tripod" in string and "cam" not in string or "grip" in string or "mirror" in string or "macro" in string or "hsm" in string or "extend" in string or "zoom" in string or "convert" in string or "ring" in string:
        return "Camera Accessories"
    elif "case" in string or "pouch" in string or "bag" in string and "cam" not in string and "lens" not in string or "cover" in string:
        return "Bags and Cases"
    elif "charger" in string or "adapt" in string:
        return "Chargers and Adapters"
    elif "card" in string or "gb" in string:
        return "Memory Cards and Graphic Cards"
    elif "tv" in string and "mount" not in string:
        return "TV"
    elif "mount" in string:
        return "Mount"
    elif "phone" in string:
        return "Phones"
    elif "project" in string:
        return "Projector"
    elif "think" in string:
        return "Laptop"
    else:
        pass
```

```python
big_df["Category"]=big_df["name"].apply(categorize)
big_df.to_excel("ABESofMAINE_trial.xlsx")
```
