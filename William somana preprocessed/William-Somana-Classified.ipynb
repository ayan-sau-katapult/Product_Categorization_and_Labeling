```python
import glob
import pandas as pd

# get data file names
path =r'C:\Users\AyanSau\Desktop\William somana preprocessed\William Somana'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_df = pd.concat(dfs, ignore_index=True)
big_df
```

```python
big_df["web-scraper-start-url"][0].split('/')[4]
```

```python
big_df["Category"]=big_df["web-scraper-start-url"].apply(lambda x:x.split('/')[4])
big_df["Sub-Category"]=big_df["web-scraper-start-url"].apply(lambda x:x.split('/')[5])
big_df
```
