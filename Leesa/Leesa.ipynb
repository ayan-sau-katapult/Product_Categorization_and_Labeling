```python
import glob
import pandas as pd

# get data file names
path =r'C:\Users\AyanSau\Desktop\Leesa\Leesa'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_df = pd.concat(dfs, ignore_index=True)
big_df
```

```python
big_df["name"]=big_df["name"].fillna(big_df["nnn"])
#big_df["name"]=big_df["name"].fillna(big_df["Name"])
big_df=big_df.drop(["nnn"],axis=1)
big_df
```

```python
def categorized(string):
    return string.split("/")[-1]

big_df["Category"]=big_df["web-scraper-start-url"].apply(categorized)
big_df
```

```python
df_temp=pd.read_excel(r"Leesa\Leesa.xlsx")
big_df = pd.concat([big_df,df_temp], ignore_index=True)
big_df
```

```python
big_df["name"]=big_df["name"].fillna(big_df["Name"])
#big_df["name"]=big_df["name"].fillna(big_df["Name"])
big_df=big_df.drop(["Name"],axis=1)
big_df
```
