```python
import pandas as pd

df=pd.read_excel("PreProcess.xlsx").drop("Unnamed: 0",axis=1)
df.head()
```

```python
df=df.fillna("None")
df.groupby("Category").count()
```

```python
def categorize(string):
    string= "".join(str(string).split()).lower()
    if "boot" in string or "kini" in string or "swim" in string or "suit" in string or "gaiter" in string or "tote" in string or "sock" in string or "slipper" in string or "sandal" in string or "buckle" in string or "short" in string or "flag" in string or "dress" in string or "man" in string or "woman" in string or "girl" in string or "boy" in string or "jacket" in string or "pant" in string or "shirt" in string or "mask" in string or "glove" in string or "strap" in string or "jacket" in string or "jersey" in string or "socks" in string or "shoe" in string or "boot" in string or "neck" in string or "brace" in string or "beanie" in string or "guard" in string or "pad" in string or "protect" in string or "hood" in string or "vest" in string or "top" in string or "tank" in string or "hat" in string or "gait" in string or "zip" in string or "pen" in string or "cap" in string or "shoe" in string or "shirt" in string or "hat" in string or "scarf" in string or "infant" in string:
        return "apparel"
    elif "chair" in string or "table" in string or "seat" in string or "beanbag" in string or "bbq" in string or "grill" in string:
        return "Furniture"
    elif "cool" in string or "mug" in string or "tumbler" in string or "bottle" in string or "sportula" in string or "coozie" in string or "open" in string or "hold" in string:
        return "Coolers + Drinkware"
    elif "tool" in string:
        return "Tools"
    elif "magazine" in string or "ammunition" in string or "shell" in string or "bullet" in string:
        return "Ammunitions"
    elif "pistol" in string or "rifle" in string and "case" not in string:
        return "Arms"
    elif "case" in string or "pouch" in string:
        return "Weapon-cases"

    else:
        pass
```

```python
df_1=df[df["Category"]!="None"]
df_2=df[df["Category"]=="None"]
```

```python
df_2["Category"]=df_2["product-card-simple-title"].apply(categorize)
df_2.groupby("Category").count()
```

```python
# Concatenate all data into one DataFrame
big_df = pd.concat([df_1,df_2], ignore_index=True)
big_df
```

```python
big_df[big_df["product-card-simple-title"]!="None"].to_excel("big_df.xlsx")
```
