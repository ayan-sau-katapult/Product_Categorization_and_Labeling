```python
import glob
import pandas as pd

# get data file names
path =r'C:\Users\AyanSau\Desktop\academy\All_Excels'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_df = pd.concat(dfs, ignore_index=True)
big_df
```

```python
big_df.to_excel("big_df2.xlsx")
```

```python
df=pd.read_excel("big_df.xlsx").drop("Unnamed: 0",axis=1)
df.head()
```

```python
df['product-card-simple-title']=df['product-card-simple-title'].fillna(df['product-card-simple-title_2'])
df['product-card-simple-title']=df['product-card-simple-title'].fillna(df['product-card-simple-title_3'])
df['product-card-simple-title']=df['product-card-simple-title'].fillna(df['product-card-simple-title_4'])

df["url"]=df["url"].fillna(df['url_0'])
df["url"]=df["url"].fillna(df['url_2'])

df["Category"]=df["Category"].fillna(df['Category_2'])
df["Category"]=df["Category"].fillna(df['Category_3'])
df["Category"]=df["Category"].fillna(df['Category_4'])
df["Category"]=df["Category"].fillna(df['Category_5'])
df["Category"]=df["Category"].fillna(df['Category_6'])
df["Category"]=df["Category"].fillna(df['category_7'])
```
