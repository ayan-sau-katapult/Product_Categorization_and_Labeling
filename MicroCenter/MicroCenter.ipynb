```python
def IS_Leasable(string):
    if string=="Back Packs and Cases" or string=="Battery" or string=="Mount" or string== "Tripod Stand":
        return "Non-Leasable"
    else:
        return "Leasable"
```

```python
df["Leasable or Non-Leasable"]=df["Category"].apply(IS_Leasable)
df.head()
```
