```python
import glob
import pandas as pd

# get data file names
path =r'C:\Users\AyanSau\Desktop\DennisKirk\All-files'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_df = pd.concat(dfs, ignore_index=True)
big_df
```

```python
big_df["web-scraper-start-url"][0].split('/')[4]
```

```python
big_df["Category"]=big_df["web-scraper-start-url"].apply(lambda x:x.split('/')[3])
big_df["Sub-Category"]=big_df["web-scraper-start-url"].apply(lambda x:x.split('/')[4])
big_df
```

```python
big_df["names"]=big_df["names"].apply(lambda x:" ".join(x.replace("\n",'').split()))
big_df
```

```python
big_df[big_df["Sub-Category"]!='100.ipp']
```

```python
df_inner = pd.merge(big_df, big_df[big_df["Sub-Category"]=="100.ipp"].drop(["Category","Sub-Category","web-scraper-start-url","web-scraper-order"],axis=1), on='names', how='inner')

df_inner[df_inner["Sub-Category"]!='100.ipp']
```

```python
df_temp=pd.concat([big_df[big_df["Sub-Category"]!='100.ipp'], df_inner[df_inner["Sub-Category"]!='100.ipp']], ignore_index=True)
df_temp
```

```python
df_temp.to_excel("DennisKirk-Combined.xlsx")
```

```python
import numpy as np

def is_leasable(string):
    if string=="nan":
        string= "".join(str(string).split()).lower()
        if 'carpet' in string or 'lining' in string or 'towel' in string or 'apparel' in string or 'mug' in string or 'sticker' in string or 'logo' in string or 'oil' in string or 'paste' in string or 'coolant' in string or "fluid" in string or 'men' in string or 'women' in string or 'man' in string or 'woman' in string or 'kid' in string or 't-shirt' in string or 'shirt' in string:
            return False
        elif 'light' in string or 'pump' in string or 'module' in string or 'computer' in string or 'sensor' in string or 'controlunit' in string or 'stereo' in string or 'audio' in string or 'radio' in string or 'sound' in string or 'cam' in string or 'camera' in string or 'aircon' in string or 'a/c' in string or 'vacuum' in string:
            return True
        elif 'chino' in string or 'mount' in string or 'service' in string or 'cover' in string or 'kit' in string:
            return False
        else:
            return False
    else:
        pass
```

```python
df["IS_Leasable"]=df["IS_Leasable"].apply(is_leasable)
df
```
