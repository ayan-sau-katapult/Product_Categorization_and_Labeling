```python
def categorize(string):
    if string=="100.ipp":
        string= "".join(str(string).split()).lower()
        if "man" in string or "woman" in string or "girl" in string or "boy" in string or "jacket" in string or "pant" in string or "shirt" in string or "mask" in string or "glove" in string or "strap" in string or "jacket" in string or "jersey" in string or "socks" in string or "shoe" in string or "boot" in string or "neck" in string or "brace" in string or "beanie" in string or "guard" in string or "pad" in string or "protect" in string or "hood" in string or "vest" in string or "top" in string or "tank" in string or "hat" in string or "gait" in string or "zip" in string or "pen" in string:
            return "apparel"
        elif "helmet" in string or "cam" in string or "goggles" in string or "visor" in string:
            return "helmets"
        elif "tire" in string or "wheel" in string or "tube" in string:
            return "tires"
        elif "exhaust" in string or "silencer" in string or "arrestor" in string or "muffler" in string or "gasket" in string:
            return "exhaust"
        else:
            pass
    else:
        return string
```

```python
df["Sub-Category"]=df["Sub-Category"].apply(categorize)
```
