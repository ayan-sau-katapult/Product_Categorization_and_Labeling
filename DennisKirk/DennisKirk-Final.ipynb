```python
import glob
import pandas as pd

# get data file names
path =r'C:\Users\AyanSau\Desktop\DennisKirk\All_parts_combined'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_df = pd.concat(dfs, ignore_index=True)
big_df
```

```python
big_df["web-scraper-start-url"][0].split('/')[4]
```

```python
big_df["Category"]=big_df["web-scraper-start-url"].apply(lambda x:x.split('/')[3])
big_df["Sub-Category"]=big_df["web-scraper-start-url"].apply(lambda x:x.split('/')[4])
big_df
```

```python
big_df["names"]=big_df["names"].apply(lambda x:" ".join(x.replace("\n",'').split()))
big_df
```

```python
big_df.groupby("Sub-Category").count()
```

```python
df_1=big_df[big_df["Sub-Category"]!="100.ipp"]
df_1
```

```python
df_2=big_df[big_df["Sub-Category"]=="100.ipp"]
df_2
```

```python
def Sub_Category(string):
    string= "".join(str(string).split()).lower()
    if "man" in string or "woman" in string or "girl" in string or "boy" in string or "jacket" in string or "pant" in string or "shirt" in string or "mask" in string or "glove" in string or "strap" in string or "jacket" in string or "jersey" in string or "socks" in string or "shoe" in string or "boot" in string or "neck" in string or "brace" in string or "beanie" in string or "guard" in string or "pad" in string or "protect" in string or "hood" in string or "vest" in string or "top" in string or "tank" in string or "hat" in string or "gait" in string or "zip" in string or "pen" in string:
            return "apparel"
    elif "helmet" in string or "cam" in string or "goggles" in string or "visor" in string:
        return "helmets"
    elif "tire" in string or "wheel" in string or "tube" in string:
        return "tires"
    elif "exhaust" in string or "silencer" in string or "arrestor" in string or "muffler" in string or "gasket" in string:
        return "exhaust"
    else:
        return "accessories"
```

```python
df_2["Sub-Category"]=df_2["names"].apply(Sub_Category)
df_2
```

```python
big_df = pd.concat([df_1,df_2], ignore_index=True)
big_df
```
