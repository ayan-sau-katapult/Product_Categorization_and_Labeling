```python
import glob
import pandas as pd

# get data file names
path =r'C:\Users\AyanSau\Desktop\DennisKirk\All-files'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_df = pd.concat(dfs, ignore_index=True)
big_df
```

```python
big_df["Category"]=big_df["web-scraper-start-url"].apply(lambda x:x.split('/')[3])
big_df["Sub-Category"]=big_df["web-scraper-start-url"].apply(lambda x:x.split('/')[4])
big_df
```

```python
big_df["names"]=big_df["names"].apply(lambda x:" ".join(x.replace("\n",'').split()))
big_df
```

```python
big_df.groupby("Sub-Category").count()
```

```python
big_df1=big_df[["names","Category","Sub-Category"]]
big_df1
```
