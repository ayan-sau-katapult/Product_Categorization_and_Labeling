```python
import glob
import pandas as pd

# get data file names
path =r'C:\Users\AyanSau\Desktop\motorola\motorola'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_df = pd.concat(dfs, ignore_index=True)
big_df
```

```python
path =r'C:\Users\AyanSau\Desktop\motorola\motorola'
filenames = glob.glob(path + "/*.xlsx")

dfs = []
for filename in filenames:
    dfs.append(pd.read_excel(filename))

# Concatenate all data into one DataFrame
big_df2 = pd.concat(dfs, ignore_index=True)
big_df2
```

```python
big_df = pd.concat([big_df,big_df2], ignore_index=True)
big_df
```

```python
big_df["href"]=big_df["href"].fillna(big_df["item-area href"])
big_df["Price"]=big_df["Price"].fillna(big_df["price"])
big_df=big_df.drop(["item-area href","price"],axis=1)
big_df
```
