```python
import glob
import os

list1=[]
empty_folders=[]
# get data file names
path =r'C:\Users\AyanSau\Desktop\tunermotor\tunermotor'
filenames = glob.glob(path + "/*.csv")

for i in filenames:
    if os.path.getsize(i)==10:
        list1.append(i.split("\\")[-1])
        empty_folders.append(i)
        
       
list1
```

```python
import pandas as pd

dfs = []
for filename in filenames:
    if filename not in empty_folders:
        dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_df = pd.concat(dfs, ignore_index=True)
big_df
```
