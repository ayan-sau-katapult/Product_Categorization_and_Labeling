```python
import glob
import pandas as pd

# get data file names
path =r'C:\Users\AyanSau\Desktop\ColemanFURNITURE\Coleman_ALLparsedFILES'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_df = pd.concat(dfs, ignore_index=True)
big_df.head()
```

```python
big_df.isnull().sum()
```

```python
big_df["name"]=big_df["name"].fillna(big_df["nnn"])
big_df["name"]=big_df["name"].fillna(big_df["Name"])
big_df["price"]=big_df["price"].fillna(big_df["ppp"])
big_df["price"]=big_df["price"].fillna(big_df["Price"])
```

```python
big_df.isnull().sum()
```

```python
big_df=big_df[["name","price","Category","Leasable"]]
big_df
```

```python
big_df.to_excel("Coleman.xlsx")
```

```python
def categorize(string):
    
    string= "".join(str(string).split()).lower()
    if "mirror" in string:
        return "Mirror"
    elif "demilune" in string or "seat" in string or "chair" in string or "bench" in string or "stool" in string or "table" in string or "bed" in string or "banquette" in string or "chest" in string or "shelf" in string or "rack" in string or "hang" in string or "cabinet" in string or "storage" in string or "desk" in string:
        return "Furniture"
    elif "pillow" in string or "throw" in string:
        return "Pillow"
    elif "mat" in string:
        return "Matress"
    else:
        return "Decorative Items"
```

```python
def is_leasable(string):
    if string=="Decorative Items" or string=="Pillow":
        return False
    else:
        return True
```

```python
big_df["Category"]=big_df["name"].apply(categorize)
big_df["Leasable"]=big_df["Category"].apply(is_leasable)
```

```python
big_df.to_excel("Coleman.xlsx")
```
